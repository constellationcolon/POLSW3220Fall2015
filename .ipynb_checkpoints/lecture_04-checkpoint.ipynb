## Individual Preferences
Require 2 things:

-   Completeness
    -   Either prefer one over the other or indifferent
    -   Unsaid: by **how much** you prefer one over the other (**not** weak preference, $R$, which just allows for indifference on top of preference as compared to $P$)
        -   Tricky to do **interpersonal comparisons of utility**

-   Transitivity
    -   Mathematically defined as
    
        $(aRb \cap bRc) \rightarrow aRc ,\, \forall a, b, c$

-   Want to rule out $\{xPy, yPz, zPx\}$.
    -   A property of triples
    -   Types
        -   PP transitivity
        -   PI transitivity
        -   IP transitivity
        -   II transitivity


## Proofs

### Proof of PP Transitivity
Want to prove $xPz$ from:

-   Assumption 1 (A1): $xPy$
-   Assumption 2 (A2): $yPz$
-   Assumption * (A*): $\sim xPz$ (Proof by contradiction)

1.  $xRy$ (A1)
2.  $yRz$ (A2)
3.  $xRz$ (1, 2, T)
4.  $xPz \cup xIz$ (3)
5.  $zRx$ (A*)
6.  $zRx \cap xRy$ (1, 5)
7.  By definition of transitivity, $zRy$
8.  (7) contradicts (A2)
9.  $\therefore \: \sim(\sim xPz)$ true$ \implies xPz$ true

### Proof of PI Transitivity
Want to prove $$(xPy \cap yIz) \rightarrow xPz$$

-   Assumption 1 (A1): $xPy$
-   Assumption 2 (A2): $yIz$
-   Assumption * (A*): $\sim xPz$

1.  $zRx$ (A*)
2.  $xRy$ (A1)
3.  $yRz$ (A2)
4.  $(yRz \cap zRx) \rightarrow yRx$ (A2, 1, T)
5.  (4) contradicts (A1)
6.  $\therefore \: \sim ( \sim xPz ) \implies xPz$

### Another example of a proof by contradiction using prime numbers
Assume $\exists$ a finite number of primes $\{p_1, p_2, \cdots, p_n\}$.

Let $S = 1 + \prod_{i=1}^n p_i$.

Is $S$ prime or composite?

$S$ can't be prime because it's not in the set $\{p_i\}$.

So if $S$ is composite, what are its prime factors? But $S \mod p_i = 1$, so it doesn't have prime factors in the finite set $\{p_i\}$ of all prime numbers. Contradiction!

$\therefore \exists$ infinite primes


## Group Preferences
Given a set $S$, we can define

-   $C(S)$, the choice set
    -   The set of alternatives x where xRy for all other alternatives y
-   $M(S)$, the maximal set
    -   The set of alternatives x where there is no y for which yPx for all other alternatives y

### Plurality rule with run-off `extends` plurality rule
For 3 candidates

Run election twice until you whittle down to 2 choices

BUT only creates _appearance_ of an absolute majority choice

### Plurality/Sequential Elimination choice `extends` plurality rule
For >3 candidates, run election multiple times until you whittle down to 2 choices.

### Round Robin Tournament aka The Method of Majority Rule (MMR) `extends` majority rule
For >2 candidates, compare 2 candidates at a time until you get a complete set of $\{c_iPc_j\}$.

### Amendment Agenda
For >2 candidates, compare 2 candidates in pairs, winner goes on to get compared with someone else.

**Outcome depends on the order of comprisons, not commutative.**

-   Agenda 1: {{x, y}, z} $\implies$ {x, z} $\implies$ z wins

-   Agenda 2: {{x, z}, y} $\implies$ {z, y} $\implies$ y wins

-   Agenda 3: {{y, z}, x} $\implies$ {y, x} $\implies$ x wins

Need to have an agenda for the ordering of agendas -.-

## The Seven Dwarves Problem
Try:

1.  Happy vs. Dopey: {1} vs. {2, 3} $\rightarrow$ Dopey wins
2.  Dopey vs. Grumpy: {3} vs. {1, 2} $\rightarrow$ Grumpy wins
3.  Grumpy vs. Sneezy: {2} vs. {1, 3} $\rightarrow$ Sneezy
4.  Sleepy
5.  Bashful
6.  Doc

Doc wins!

BUT.

-   Doc is not very high on anybody's list.
-   Happy is quite highly ranked, but didn't win
-   Latecomer advantage
