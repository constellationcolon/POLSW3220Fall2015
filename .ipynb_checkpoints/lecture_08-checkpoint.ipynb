## Last Time
Conditions U, P, I, T

### IIA
When comparing 2 alternatives, ignore all other alternatives

Don't care about degree of preference of one over the other

If in one profile x is chosen over y, then in another profile with the same profile on the x-y pair level, we have to choose x over y in the other profle too

### Dictator
Condition D: Don't have an **Arrowian Dictator**

Someone who gets what they want no matter what everyone else wants in every pairwise comparison no matter what everyone else wants

```python
x ----------------------- o
lone                 spread
power              of power
```

Arrow's condition: don't be the x.

If you want U, P, I, D, you cannot guarantee transitivity

You **can** get any 4 of the 5, but not all 5 at once.

Question: Are we going to place any constraints on the output?


## Social Welfare Function (SWF)
Constraint on output: group choice must be transitive

Also:

*   Social Choice Function
*   Social Decision Function

### Arrow's Theorem (redux)
There is no social welfare function that satisfies all of U, P, I, and D

### Examples
*   _Is there a SWF that satisfies U, P, and I?_
    *    Just establish an Arrowian Dictator.

*   _SWF that satisfies U, P, and D?_
    *   Borda count. Plurality rule?

*   _SWF that satisfies U, I, and D?_
    *   Alphabetical order. Imposition from the outside that overrides internal preferences. **Your preferences don't matter.**

*   _SWF that satisfies P, I, and D?_
    *    Unanimity rule.

*   _U, P, I, and D without the SWF (i.e. violate T)?_
    *   Method of Majority rule
        *   U: for every pair, we get a choice
        *   P: If there is unanimous preference, then duh, choose that
        *   D: If the majority is opposed to a dictator, then there **is** no dictator under MMR
        *   I: By definition, this is satisfied, since we vote by pairwise comparisons
    *   P and $\neg$D are compatible since the dictator is part of the group that's giving rise to the unanimity


### Implications of Arrow's Theorem

1.  What should we give up on, then? Should we go back to majority rule?
2.  What exactly is Arrow's Theorem saying?
    *   Tradeoff between social rationality vs the concentration of power
    *   "Logic" vs "fairness"
