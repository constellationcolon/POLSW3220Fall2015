## Transitivity, Quasi-Transitivity, Acyclicity

### Transitivity
Need to satisfy all of RR, PP, IP, PI, II

### Quasi-Transitivity
Need to satisfy PP only

*   e.g. aIb, aIc
    *   Transitivity: only bIc allowed
    *   Quasi-Transitivity: bIc, bPc, cPb all allowed

### Acyclicity
Triple Acyclicity not enough. Must look for chains of every length.

### Logic
*   Conditions of the form $p \rightarrow q$
    *   What happens when $p$ is false? $p \rightarrow q$ automatically true.
    *   e.g. if you don't have a chain that would violate T, then T is not violated.

## Pareto Extension Rule
Strictly choose an alternative ($xPy$) $\iff$ somebody prefers that ($xP_iy$) and no one says otherwise ($\not\exists yP_ix$)

If there is any conflict at all, then xIy.

Sen 1969 Pareto Extension Rule: satisfies UPID.

$$xRy \leftrightarrow \neg( \forall i, y R_i x \cap \exists_k yP_k x ) $$

Even satisfies strong versions of P (weak pareto condition) and D, P' (strong pareto condition) and D' (strong non-dictatorship condition).

## Strong Pareto Condition (P')
If there are people who $xP_iy$ and the others $xIy$, then you have to choose $xPy$.

## Strong Non-Dictatorship Condition (D')
### Weak
$$ \not \exists i: \forall x, y, ~ xP_iy \rightarrow xPy $$

Only rules out the strongest of dictators

### Strong
$$ \not \exists i, x, y, ~ xP_iy \rightarrow xPy $$

Rules out weaker dictators in addition to strong dictators

How to be a weak(er) dictator (than an Arrowian Dictator)

1.  Local dictator over a particular pair: $\exists i, x, y, ~ xP_iy \rightarrow xPy$
2.  Weak preference dictator: $xR_iy \rightarrow xRy$

## Theorems
### 0. Arrow's Theorem
SWF with UPI $\implies ~ \neg$D 

### 1. Sen (1969) Theorem
Pareto Extension Rule

### 2. Gibberd's (1969) Theorem
QT SDF with UPI $\implies$ (Global) Oligarchy

#### Oligarchy
(improper) subset of people who have disporportionate power over all pairs of alternatives

Identifiable and unique group that together has dictator-like powers over all pairs of alternatives

Only way to have an oligarchy and not violate anonymity, is to make everyone an oligarch

Not every violation of anonymity creates an oligarchy. Could have just special powers over local pairs.

Impossible to have more than one oligarchy

Only way for entire group to be in oligarchy is the Pareto Extension Rule

##### Veto power
Individual power

$$(\exists i \in \text{oligarchy}: \forall x,y ~~ xP_iy) \rightarrow (xRy \equiv \neg yPx)$$

##### Power to impose
Collective power

$$(\forall i \in \text{oligarchy}: \forall x,y xP_iy) \rightarrow xPy$$

In the Pareto Extension Rule, everyone is an oligarch.

### 3. Brown's (1975) Theorem
If $N(\text{alternatives} \geq N(\text{individuals}))$, SDF UPI $\implies$ Collegium

Just want acyclicity with UPI, still have to create a special group called collegium.

Want more group rationality $\implies$ Need to concentrate group power

#### Decisive Coalition
When unanimous in strict preference $xPy$, imposes at the group level that strict preference for $xPy$

$$(\forall i \in \text{DC}, xP_iy) \rightarrow xPy$$

Not special!

The entire group is always a decisive coalition

Any group of any size that can always get what it wants

Are there people who can be in every possible decisive coalition?

*   Not under majority rule
*   But sometimes yes!

#### Collegium
You are in the collegium if you are in every decisive collegium!

Oligarchy $\subset$ Collegium

If there exists someone with veto powers, then that person is in the collegium. BUT you don't need veto powers to be in the collegium.

Most of the time when a choice rule involves veto powers, there is a collegium. But there is no logical requirement that anyone in the collegium has veto powers.

#### e.g. UN Security Council
To pass: need majority vote, no veto from security council

Security council has no power to impose, but has veto powers

Security council is in the collegium

But don't need to have veto powers to be in the collegium. Just need to be in the intersection of all decisive coalitions.
