## Liberalism (L)

There are some choices in which individuals are going to get into trouble

Want individuals to be decisive over certain choices

### Sen's Liberalism

Every individual has at least one pair of alternatives that they get to control

### Minimal Liberalism

Only 2 individuals have Sen's Liberalism

### Example: Lady Chatterley's Lover (1928)

Sexually frank book, highly controversial

Only one copy of book, only one can read it at a time

Person A: prude, Person B: lewd

3 states: $\{$ $X_A$: Only A reads, $X_B$: Only B reads, $X_0$: Neither reads $\}$

Prude's (A's) preferences: $X_0X_AX_B$

Lewd's (B's) preferences: $X_AX_BX_0$

A should get to decide between $X_A$ and $X_0$

B should get to decide between $X_B$ and $X_0$

Between $X_0$ and $X_A$, $A$ is decisive, so group should choose $X_0$

Between $X_0$ and $X_B$, $B$ is decisive, so group should choose $X_B$

Given these, how to satisfy quasi-transitivity? $X_BPX_A$? But $X_APX_B$ by Pareto Condition. There will be a cycle in trying to decide between $X_A$ and $X_B$.

e.g. Can't choose $X_B$ (since we think that we can just let the lewd read it!) since we can make both people happier by shifting to $X_A$.

$\not\exists$ SDF that satisfies U, P, and ML (and certainly not L!) when there are $\geq$ 2 people

### aka Impossibility of the Paretian Liberal

How to get out of this paradox?

Tension between Pareto condition and Liberalism

How do we know when to ignore Person A? Troubling because they have "nosiness": preferences that concern what other people do.

So should we always ignore "nosy" preferences? Should ignore when it doesn't directly affect them? But hard to define what is "directly affect". Difficult to draw the line; e.g. Halloween costume controversy in Yale.

Resolve via sequential choice? Both have right to read the book but can only read one at a time.

### Example: Trial by Jury

Does motive matter? Or only the outcome?

Angelina, Edwin, Judge

Angelina suing Edwin for leaving her at the altar and eloping with bridesmaids, Judge resolves it by just marrying Angelina himself.

Alternatives:
$\{$
$X_S$: stay single,
$X_J$: marry Judge.
$X_E$: marry Edwin
$\}$

Preferences:
$\{$
$A: X_EX_JX_S$,
$E: X_SX_EX_J$,
$J: X_EX_JX_S$
$\}$

$E$ decisive over $X_S, X_E$;
$A$ decisive over $X_S, X_J$

Why can't we choose $X_E$? Well, Edwin is decisive over $X_E$ and $X_S$, and he would prefer $X_S$.

Why not $X_S$? Angelina decisive over $X_S$, and would prefer $X_J$.

Why not $X_J$? Everyone $X_EPX_S$, so by Pareto Condition, we have to choose $X_E$.

ML violates A.

Pareto Condition seems to make sense overall, but in this conext leads to unnatural impositions on choices.

Social choice methods may seem nice but can create trouble of their own.

### Strategic behavior

**Strategic voters** will take into consideration what others will vote, then vote in such a way that they will get what they want after calculation

**Sincere voters** just hand in their preferences and count on the machine to output group choice (baseline); "faith-ordering"

**Sophisticated voters**, only with respect to fixed agenda. You may be a sophisticated voter, but there may be situations in which you'd just be sincere. Other times, will cast strategic vote.
