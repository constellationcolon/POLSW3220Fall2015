###A $\rightarrow$ B
"A is sufficient for B" or "B is necessary for A"

Anything can be implied by a false statement. If A false, B can be anything and the whole thing can still be true.

###A $\leftrightarrow$ B
"A is necessary and sufficient for B"

###A XOR B
(A $\cup$ B) $\cap$ ~(A $\cap$ B)

###Logical Equivalence
Proof by truth table

### e.g. Assume "(p $\rightarrow$ q) $\rightarrow$ happy" is true
If p is false, then p $\rightarrow$ q is true, so happy must be true.

If p true, q true, then p $\rightarrow$ q is true, so happy must be true.

If p true, q false, then p $\rightarrow$ q is false, so happy is unknown.


##Preference Relations

### Preference
$P_i$ is a strict preference (i.e. $>$ not $\geq$), NOT a _strong_ preference

-   $x P_i y$
    -    Person i prefers x to y

-   $y P_i x$
    -    Person i prefers y to x

-   ~$y P_i x$
    -   Person i does not prefer y to x

"Weak" preference: $\geq$ not $>$

-   $x R_i y \leftrightarrow (xP_iy)\cup(xI_iy)$

### Indifference

$x I_i y \equiv yI_ix$

Person is indifferent between x and y

### Completeness / Comparability
For each pair of outcomes, you have a preference or you're indifferent.

A property that operates at the level of **pairs**.

### Transitivity
$xPy \cap yPz \cap zPx$? Logically impossible.

-   PP Transitivity
    -   $xPy \cap yPz \implies xPz$

-   PI Transitivity
    -   $xPy \cap yIz \implies xPz$

-   IP Transitivity
    -   $xIy \cap yPz \implies xPz$

-   II Transitivity
    -   $xIy \cap yIz \implies xIz$

A property that operates at the level of **triples**.

#### Formal definition of Transitivity

$(aRb \cap bRc) \rightarrow aRc \:\forall$ $a, b, c$

#### RR $\rightarrow$ PP

$xRy \cap yRz \implies xRz$

Want to prove $xPz$

Assumption (1): $xPz$, Assumption (2): $yPz$
