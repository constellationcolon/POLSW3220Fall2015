## The Borda Count
-   Weighting preferences to put them into an algorithm
-   Not taking alternatives pairwise, but taking them as a whole
-   Just a convention to number the points $n \in \mathbb{Z}$

### Pros and Cons
-   Cons
    -   Takes that count very seriously
    -   Easy misconception: I prefer A 11 times over B
    -   Acting as though the count is something more than just a ranking
    -   Adding information and pretending that we have more information than we actually do have
    -   **Not the case that a Condorcet winner will necessarily win**
-   Pro
    -   Prevents Condorcet Losers from winning (Condorcet-Loser Proof)

There are methods that guarantee that Condorcet Winner will win (e.g. using an agenda). These methods are called **Condorcet Consistent**. Conversely, those that automatically kick out Condorcet Losers are called **Condorcet-Loser Proof**

## Copeland Rule
-   Victory **counts**
-   Take an alternative a, compare to another x. If a wins, +1 to a. If a loses, -1 to a. If tie, 0 change.
-   Given n alternatives, can get max n - 1 points (for Condorcet Winner)
-   Condorcet consistent, Condorcet-loser proof

### 1776 Example
-   a vs: b: -1, c: -1 = -1
-   b vs: a: +1, c: -1 = 0
-   c vs: a: +1, b: +1 = 2


## Simpson Rule
-   Victory **sizes/margins**
-   Compute for each x the number of people who prefer a to x
-   Simpson score is the **minimum** of the victories score
-   Condorcet consistent, not Condorcet-loser proof
    -   Condorcet winner will have score of $> \frac{n_{\text{people}}}{2}$

### 1776 Example
-   a vs: b: 8, c: 8 => 8
-   b vs: a: 13, c: 8 => 8
-   c vs: a: 13, b: 13 => 13

- Tie between a and b, c is the winner
- Tests for "weak spots"

## Recall: Plurality/Sequential Elimination
-   Vote across all alternatives, eliminate the weakest alternative, vote again with the rest, repeat until you get a winner

## Nanson's Rule (aka Borda Elimination)
-   Use Borda Count on Plurality Elimination
-   Each round, drop the alternative with the lowest Borda Count
       - Remember to renumber the grid because the scores are different
-   Condorcet Consistent and Condorcet-Loser Proof
    -   CLP because in the last step, you always have to go up against someone else... who will always win against the CL

### 1776 Example
- Round 1
    - a: 37
    - b: 42
    - c: 37
- Round 2
    - b: 29
    - c: 34
- c is the winner
- When it comes down to 1v1, Nanson's is the same as majority rule, just 2 points to 1 as opposed to 1 point to 0 in majority

## Desirable Properties for Rules
### 1. Universal Admissibility (U)
-   Algorithm should work regardless of your preferences
-   Preferences -> Blackbox -> Output
-   Blackbox shouldn't explode just because you have a certain set of preferences
-   Must take into account all cases (e.g. what happens if there's a tie?)
-   Recall: "Preference Profile" is a set of individuals and their preferences
-   Sometimes we want winner, sometimes we just want group ordering
    -   Choice set vs. Maximal Set
        -   Choice set X: $x \in C(X) \iff \forall y,\, xRy$
            -   x must beat every other alternative
            -   Only pairwise comparisons, so might be a problem
            -   Must at least be weakly preferred over everything else
            -   NOT the definition of Condorcet winner. Condorcet winner is winner in **majority vote**. Here we have pairwise comparisons, regardless of the group rule used.
        -   Maximal set X: $x \in M(X) \iff \not\exists y, yPx$
            -   There is nothing that strictly beats $x \in X$
    -   Can almost always use them interchangeably
    -   If there is completeness, then choice set $\equiv$ maximal set
    -   e.g. { xPy, xIz }
        -   C(X) = { x }
        -   M(X) = { x, z }
        -   neglecting transitivity
        -   z is not the C(X) since there is no comparison between y and z

### 2. Pareto Principle (P)
-   If everyone agrees that x is better than y, then we should choose x over y
-   $\forall i,\, xP_iy \rightarrow xPy$

### 3. Independence from Irrelevant Alternatives and Irrelevant Information(IIA)
-   e.g.
    - GNB 44%
    - BGN 47%
    - NGB 9%
-   For any given pair of alternatives, who wins is not just determined by preferences of those 2, but also by the rankings of other alternatives
-   Under IIA, the only preferences that are allowed to matter in the group ranking are the individual preferences for those 2 alternatives.
-   Not meaningful to compare degree of preference. I don't want to know by how much you prefer x to y or y to x.
-   Not meaningful to compare irrelevant alternatives z or m or j when you are just comparing x and y
-   If there are 2 profiles, and if at the level of the x-y comparison the 2 profiles are identical, then IIA makes sure that the outcome is the same
-   IIA violated if the 2 profiles are identical when comparing that specific pair and group choice is not the same
-   If 2 profiles are the same on the pair level, then the group choice has to be the same in both.
-   Depend only on the individual level of how the alternatives are ranked.
