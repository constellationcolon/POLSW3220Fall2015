## May's Theorem

### How to tell if anonymity was violated?
If a preference profile has the same sets of preference orderings, but shuffled around between people, and the group outcomes are different, then anonymity has been violated

### Neutrality
*   Arrow's indifference condition: IIA
*   I+
    *   Every pair comparison has to be treated independently, completely neutrally
    *   What you're allowed to take into account when doing the paired comparison
*   How to show neutrality?
    *   Show that x-y comparisons were done differently across profiles
    *   Show that x-y comparisons were done differently from w-z comparisons within the same profile if the paired preference orderings can be mapped exactly

### Monotonicity
*   People's preferences must have a **positive** impact on the group ordering
*   If at the individual level nobody has moved towards y and at least 1 has moved towards x, then group preference should move towards x, not towards y
    *   If the group was previously indifferent between x and y, then the group must then now prefer x to y

### The Theorem
Method of Majority Rule $\iff$ UANM

If you don't think MMR is the way to go, then you should be able to phrase why you aren't going to use it in terms of UANM.

e.g. Don't want to use MMR because of cycling, then that means you prize no cycling over at least one of UANM

e.g. Why don't you want to use MMR for the death penalty? Because you don't want to treat death vs. no death equally (don't want N)

e.g. Why don't you want to use MMR for the presidential veto? Because you want the president to be treated as anonymous (don't want A)

## May-Arrow Theorem (aka the No Name Theorem)
UANM vs UPID

*   U $\iff$ U
*   A $\implies$ D
*   N $\implies$ I
*   $\neg ($M$ \implies $P$)$
*   (M$\cap$N) $\implies$ P

### Proof of $ ( M \cap N ) \implies P $ (by contradiction)
Assume M, N, $\neg$ P

$\neg$ P: 100% $xP_iy$ (individual) but $yPx$ (group)

shift to 99% $xP_iy$, 1% $yP_ix$(individual) $\implies$ $yPx$ (group) because M

shift to 50% $xP_iy$, 50% $yP_ix$(individual) $\implies$ $yPx$ (group) because M

shift to 0% $xP_iy$, 100% $yP_ix$(individual) $\implies$ $yPx$ (group) because M

Contradiction: Neutrality requires that 100% $yP_ix$(individual), $xPy$ (group) or else we are not treating the x-y comparison the same way we are treating the y-x comparison

$\therefore \neg$($\neg$P) $\implies$ P

### Real Life Example
1980: Virginia State Senate, deciding on the ERA (Equal Rights Amendment)

Looked like it was going to be tied 20 - 20 in the 40 member senate

Rule 1: Lieutenant Governor gets to pass the tie-breaking vote (Chuck Robb, ERA supporter)

Rule 2: Needs to have an absolute majority for something to pass (i.e. 21 votes)

One of the senators, John Chichester, opponent of ERA:

*   If he votes no, 20 - 20, passes 21 - 20 by the tiebreaking vote
*   John abstained; allowed in circumstances of "immediate private or personal interest"—John cited "interest in the wellbeing of Virginia"
*   Final vote: 20 - 19 $\implies$ no tie, votes never get to 21, because no tie-breaking vote $\implies$ failed

Does not satisfy monotonicity. John moved his preferences towards being for the ERA (against $\rightarrow$ abstain), but the group outcome moved towards being against the ERA.

Does not satisfy anonymity, because Chuck Robb has tie-breaking power. Does he have more or less power?

*   Even voters without the tie breaker
    *   Irrelevant if there is no tie. If there is no tie, it's as if he voted
*   Odd voters without the tie breaker
    *   Can make sure there is a tie if there is no tie.

Does not satisfy neutrality. Easier to not pass something than to pass something. Can get something to fail with less votes.

## Paradoxes
Violate M or N

### Example: Additional Support Paradox

#### Profile A:

```
6   5   4   2   <- Number of voters
=============
a   c   b   b   <- Preference orderings
b   a   c   a
c   b   a   c
```

Use plurality rule with run-off:

*   a: 6, b: 6, c: 5
*   a: 11, b: 6

$\implies$ aPbPc

### Profile A':

```
6   5   4   2
=============
a   c   b   a
b   a   c   b
c   b   a   c
```

Same profile, only last column different

Use plurality rule with run-off:

*   a: 8, b: 4, c: 5
*   a: 8, c: 9

$\implies$ cPaPb

#### Contradictions
*   Violates monotonicity (not May's M, but a weaker version)
    *   moved $a$ up in ranking, but $a$ did worse
    *   Additional support paradox
    *   $cP(aIb)$
*   Violates neutrality

### Example: No Show Paradox

Occur in systems that violate N or M

#### Profile 1

```
I   II  III IV
26  47  2   25
a   b   b   c
b   c   c   a
c   a   a   b
```

Plurality rule with run-off:

*   a: 26, b: 49, c: 25
    *   strike c off
*   a: 51, b: 49

aPbPc

Group II gets their last choice

#### Profile 2

What happens if group II oversleeps and misses the vote entirely?

```
I   III IV
26  2   25
a   b   c
b   c   a
c   a   b
```

*   a: 26, b: 2, c: 25
    *   strike b off
*   a: 26, c: 27

cPaPb

Group II gets their first choice by not voting instead!

Also can view no show as indifference. i.e. show $\rightarrow$ no-show : : move towards a, away from c

*   Violates M


## Approval Voting
*   Borda count sensitive to ranking
*   Plurality only sensitive to top ranking

Method: For each alternative, just say yes or no.

Has an "approval line"

```
I   II  III IV
26  47  2   25
a   b   b   c   Approve => 1 point
b   —   c   a   
—   c   —   —   
c   a   a   b   Disapprove => 0 points
```

*   a: 26 + 25 = 51
*   b: 26 + 47 + 2 = 75
*   c: 2 + 25 = 27

Don't need to make harsh choices between 1st and 2nd like in plurality rule
