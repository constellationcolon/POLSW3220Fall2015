## Transitivity
*   Transitivity
    *   Sufficient for Choice Set
    *   Full RR transitivity
        *   PP, PI, IP, and II
    *   **Social Welfare Function**
    *   On the level of triples

*   Quasi-transitivity
    *   Sufficient for Choice Set
    *   Requires only PP
    *   $(aPb \cap bPc) \rightarrow aPc, ~ \forall a, b, c$
    *   Quasi-transitive **Social Decision Function**
    *   On the level of triples

*   Acyclicity
    *   Necessary and Sufficient for Choice Set
    *   Social Decision Function
        *   just want a Maximal Set (if there is completeness, also the Choice Set)
    *   $ (aPb \cap bPc \cap \ldots \cap yPn) \rightarrow ( aRn \equiv \neg nPa ) ~ \forall a, b, \ldots n $
    *   Just want to eliminate a cycle of strict preferences
    *   NOT defined on the level of triples
        *   Example
            *   Suppose we have a lesser property "triple acyclicity" below acyclicity that is not sufficient to give a choice set
            *   Necessary, not sufficient for Choice Set
            *   $(aPb \cap bPc \cap bPc) \rightarrow ( aRc \equiv \neg cPa ) ~ \forall a, b, c$
            *   Consider $wPx$, $xPy$, $yPz$, $zPw$, $xIz$, $wIy$
                *   Is triple acyclicity satisfied?
                *   There is not triple that cycles
                *   But it does **not** satisfy acyclicity!
    *   To check acyclicity, have to check all the chains of strict preference, of every length
    *   **Need acyclicity to have a choice set**

N.B. Transitivity $\rightarrow$ Quasi-transitivity $\rightarrow$ Acyclicity

*   Examples
    1.  $xPy$, $yIz$, $xIz$
        *   satisfies PP (i.e. QT), but not the rest of the conditions
        *   choice set: $\{x, z\}$
    2.  $xPy$, $yPz$, $xIz$
        *   violates T and QT
        *   still have a choice set: $\{x\}$
    3.  If $xIy$, $yIz$ do the following violate QT?
        *   $xPz$? No
        *   $zPx$ No
        *   $xPz$ No
    4.  If $wPy$, $yIz$, what is compatible with QT?
        *   $wIz$, $wPz$
        *   NOT $zPw$
            *   violates PP transitivity. $zPw$ and $wPy$ requires $zPy$

## Pareto Extension Rule

### Definition
$$
xRy \leftrightarrow ( \exists i: xP_iy ) \cup (\not\exists k: yP_kx)
$$

"A really shitty choice rule! If it's any conflict at all, the group must be indifferent."

Produces so much indifference that it violates the "knife-edge" condition (Monotonicity)
