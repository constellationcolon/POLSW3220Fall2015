## Theorems

### 4. Hanson's (1969) Theorem

The only SWF (T) that satisfies neutrality (N) and anonymity (A) is **complete social indifference**.

Meaning group indifferent to all alternatives.

We would need to violate P and M.

T, N, A $\rightarrow$ $\neg$ P, $\neg$ M, $\forall x, y ~ xIy$ 

**Question** What's in the choice set if we have complete indifference? Everything!

### 5. Muller-Satterthwaite's Theorem

What happens if we want a single winner (e.g. one president)?

The only single winner SDF (QT) that satisfies U, P, I, is dictatorship.

single winner QT, U, P, I $\leftrightarrow$ $\neg$ D

### 6. Sen's (1977a) Theorem

If the number of alternatives is greater than or equal to 4, and you want SDF, as well as to satisfy UPIM, then someone has veto power.

All you need in general for SDF is acyclicity for a choice set to exist.

$(N(x) \geq 4 \cap QT \cap UPIM) \rightarrow (\exists i: \forall x, y ~ xP_iy \rightarrow xRy)$

Econometric theorems by Ma-Colell & Sonnenschen in 1972, etc.

### 7. Black's "Not Worst" Theorem

"Not worst" is a property of triples. If you give me 4 alternatives, I need to look at them one triple at a time.

For any triple I have a best, middle, worst. You can ask if anything is "not worst".

It gets trickier when dealing with indifferences, but for now let's focus on strict preferences.

For an alternative to be "not worst", ask if everyone can agree that one of them is not worst.

#### Example

Everyone agrees that amongst $\{w, x, y, z\}$,

*   $y$ is not worst in $\{xyz\}$
*   $y$ is not worst in $\{wxy\}$
*   $z$ is not worst in $\{wxz\}$
*   $z$ is not worst in $\{wyz\}$

We won't get cycling if there is agreement. If for every triple there is a "not worst", cycling will not occur.

If the number of people is odd, we are guaranteed transitivity. If the number of people is even, we are guaranteed at the very least quasi-transitivity.

Doesn't tell us what we chould choose, just tells us if there will be cycling.

If we satisfy Black's not worst condition, then the MMR will not cycle. There isn't one preference on its own that can cause cycling.
