## Value Restriction Theorem

Given $\{x,y,z\}$, group agrees that $x$ is not best/mid/worst $\implies$ satisfies **value restriction (VR)** for that triple

### Example

Given $w,x,y,z$, if unanimously agree that

*   In $\{w,x,y\}$, $y$ is not mid
*   In $\{w,x,z\}$, $z$ is not best
*   In $\{w,y,z\}$, $y$ is not worst
*   In $\{x,y,z\}$, $x$ is not worst

Then we are guaranteed no cycling

### Definition

In a triple $\{x,y,z\}$, $x$ satisfies VR if

$$\underset{x \text{ not worst}}{[\forall i: xP_iy \cup xP_iz]} \cup \underset{x \text{ not best}}{[\forall i: yP_ix \cup zP_ix]} \cup \underset{x \text{ not mid}}{[\forall i: (xP_iy \cap xP_iz) \cup (yP_ix \cap zP_ix)]}$$

If all orderings are **strict**, then VR is **sufficient and necessary*** for majority rule to work well

*   **Necessary*** necessary, unless you let me play with the distribution
*   If necessary* not satisfied, then there will be some distribution where cycling will occur
*   If you aren't allowed to play with the distribution, then you need to know the exact distribution to be able to decide

Always the case that if you let me move people across types, then majority rule would work well.

### Indifference

Gets a little more complicated when we allow for people to have indifference in their preferences.

Divide people into two types with regard to a given triple.

**Concerned** if you have a preference within that triple.

**Unconcerned** if you have no preferences within that triple. Completely indifferent between all three alternatives. We can ignore this group when comparing the three.

If there is an even number of people that have preferences within that triple (i.e. concerned), not guaranteed full transitivity, may only have quasi-transitivity.

We can still have agreements on not best/worst. Agreeing on not mid is harder. e.g. if you have $xIyPz$, can say that z is not best, but cannot say that $x$ or $y$ is not mid. **To say that something is not mid, an alternative must either strictly beat or strictly lose to both other alternatives in the triple.**

### Questions on VR

If I say that in a triple, $y$ is mid, is VR satisfied? Yes

If in a group of 33 alternatives that satisfies VR, and we take out $x$, is VR still satisfied? Yes

If there are 2 groups of people, and VR satisfied individually for each group of people, when we combine then together, is VR still satisfied? No


## Limited Agreement (LA)

### Definition

**Limited Agreement** is satisfied for an ordered pair $\{x,y\}$ in a triple $\{x,y,z\}$ $\iff [ \forall i: xR_iy] $

Not looking for restriction, looking for limited agreement.

Given a triple, must unanimously agree that some $x$ weakly beats some $y$

If $\forall$ triples LA is satisfied, then QT is guaranteed.


## Extremal Restriction (ER)

### Historical Perspective

#### Echoic Preferences

If anyone had strict preferences $xPyPz$, no one can go back to the reverse $zPyPx$

If no one has strict preference, then echoic preferences satisfied

#### Antagonistic Preferences

If someone has strict ordering $xPyPz$, then everyone else better prefer $xPyPz$, $zPyPx$, or $xIyIz$

#### Dichotomous Preferences

Would work well if everyone is a little bit indifference

If every individual was indifferent between at least one pair in each triple

### Definition

If someone in the group has strict preferences over a triple, then there must be some conditions that must be satisfied for majority rule to work well

```
If someone has strict ordering ordering xyz
{
    If someone else thinks zyx
    {
        we can satisfy ER
    }
    else if someone else thinks something other than xyz or zyx
    {
        then ER not satisfied
    }
}
else
{
    ER satisfied
}
```

Cannot have partial jerks (opponent). Need no jerks (opponent), or complete jerks (opponent).

**If ER is satisfied, we have transitivity.**
