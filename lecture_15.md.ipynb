## Subset Rationality

So far we've been treating profiles as fixed. What happens if we expand/decrease number of alternatives? Different scopes of competition.

### Sen's $\alpha$

$S_1 \subset S_2: x \in S_1 \implies x \in S_2$

Suppose $x \in C(S_2)$, $C$ is the choice set.

Subset rationality condition $\alpha$ (contraction rationality): $x \in S_1 \rightarrow x \in C(S_1)$

### Sen's $\beta$

$S_3 \subset S_4$; $x, y, w, z \in S_3$; $a, b, c, d \in S_4-S_3$

Suppose $x \in C(S_3), y \in C(S_3), x, y \in C(S_3)$.

Subset rationality condition $\beta$ (expansion rationality): $x \in C(S_4) \leftrightarrow y \in C(S_4)$


## Binary Relation underlies choice set

All SDFs (acyclicity) satisfy $\alpha$, but not necessarily $\beta$.

$\beta \leftrightarrow$ SWF (transitivity)

### Another way of stating Arrow's Theorem

No SDF that satisfies U, P, I, D, and $\beta$

### Connection to Sen's Conditions

Satisfy any of ER, LA, or VR, can satisfy $\alpha$

To satisfy $\beta$, have to satisfy ER
