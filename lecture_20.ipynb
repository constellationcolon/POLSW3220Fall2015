## Last Time

* Single peaked preferences
    * By majority vote, median voter's ideal point is group's (only stable) ideal point

### Jargon

* individual win set
    * set of points strictly preferred to any given point
    * the only point with a null win set is your ideal point
* group win set
    * set of points such that a majority prefers them to any given point
    * at median voter's ideal point, the win set is empty
* core
    * points at which win set is empty is called the core
    * if there is no core, then there is no stable point under majority rule
    * odd number of voters, core is a single point
    * even number of voters, core is a closed interval

## Plott's Theorem

Extension of Black's median voter theorem to higher dimensions

**Radial symmetry** Refer to `20_4.png`. $A$ and $C$ are a radially symmetric pair around $B$ because they are on the same line as $B$, _even though they are not the same distance from $B$_

What happens to other voters? They also have to be radially symmetric around $B$.

If there are odd voters, and one of the voter's position is such that the other voters are paired in radial symmetry, then central point is going to be the core.

![](20_5.png)

Radial symmetry is incredibly hard to acheive for odd numbers of voters

Even number of voters $\implies$ a tiny bit easier to acheive radial symmetry. (e.g. remove $B$ from the group of voters, then the point where $B$ used to be will be in the core)

Not very helpful in the real world.

What happens when there isn't actually radial symmetry? (which is most of the time)

## McKelvey's Chaos Theorem

### Part I

In multiple dimensions, unless there is radial symmetry, or you get very very lucky with the voter distribution, then the core is going to be **empty**.

There is not stable point $\implies$ cycling is going to occur!

### Part II

There is an amendment agenda (via sincere, majority votes) from any point you name that will get you to to any other point in the entire space.

All the points in the entire space are involved in a huge cycle.
