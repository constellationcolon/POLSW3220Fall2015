## Cycling
*   How often does cycling occur? When? How? Probabilistically?

*   Definitions
    *   Preference Profile
        *   A fixed group of people and their preferences
    *   Condorcet winner
        *   An alternative that can beat each and every other alternative in a pairwise vote
        *   Cannot be a cycle for the top choice, but _still can have a cycle over the rest of the positions_
    *   Condorcet loser
        *   An alternative that will lose to each and every other alternative in a pairwise vote
    *   Voting cycle
        *   When by majority vote, there is loop of alternatives where the first beats the second, the second beats the third, etc., and the last beats the first

*   Consider 3 people and 3 alternatives
    *   Each person has 13 possible preferences
    *   216 distinct profiles
        *   12 without a Condorcet winner
        *   204 there is a Condorcet winner
    *   Cycles are probabilistically not that common
    *   BUT if voting many times, and have many alternatives, cycles more likely
    *   3 voters, 3 alternatives: 6%
    *   5 voters, 5 alternatives: ~20%
    *   Approach to infinity: 100%

*   Hard to see cycles in the real world
    *   Often doesn't exist a well defined rank of preferences
    *   There might be dropoff to a 1v1, one from each party
    *   Preferences based on who is still running
    *   **We have a specific voting agenda in stuff like the presidential vote**
        *   This agenda prevents us from seeing the raw preferences of people
        *   Heavily censored by institutions and strategic choice

*   If there is a voting agenda and a Condorcet winner, the voting agenda will find the Condorcet winner for you
    *   Not the case that anything that wins a voting agenda is the Condorcet winner
    *   Condorcet loser can never win a voting agenda
        *   A voting agenda is guaranteed to knock out any Condorcet loser
    *   If you're first on the agenda and still win the agenda, you're a Condorcet winner

## Lawyers Problem
*   Setting
    *   10 junior lawyers in a law firm with a bunch of partners
    *   The partners are not promoting anyone to partner
    *   New rule for promotion
        *   Partners rank each junior lawyer privately
        *   Each junior knows their rank
        *   10 junior lawyers decide what the cutoff will be
*   Higher ranked junior
    *   Will prefer to have fewer fellow partners
    *   Reputation
    *   More money?
*   Lower ranked junior
    *   Will prefer no one else makes partner so that no one knows you're the worst
*   Cutoffs:
    *   10
        *   Passes 10 v 0
    *   9
        *   Passes 9 v 1
    *   8
        *   Passes 9 v 1
        *   The no vote comes from rank 9
            *   Rank 10 is glad they're not the only loser
            *   Others glad that there are less partners
    *   7
        *   Passes 9 v 1
    *   6 ... 1
        *   All pass 9 v 1
    *   0
        *   Passes 9 v 1
    *   Cycle back to 10
        *   Passes 10 v 0
*   Another example of cycling
*   Probabilistic lens misses the fact that certain settings guarantee/encourage cycling
